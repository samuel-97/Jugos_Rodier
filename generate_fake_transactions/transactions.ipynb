```python
# import libraries: 

import pandas as pd
import random
import numpy as np
```

```python
# read in menu items csvs: 

menu = pd.read_csv("~/repos/jugos_rodier/csvs/menu_items.csv")

menu = menu[['item_id', 'price', 'item_category']]

menu.head()
```

```python
# combine item_id and price into a single string: 

    # first convert price into a string:

menu['price'] = menu['price'].astype('str')

    # concatenate them together: 

menu['item_price'] = menu['item_id'] + '_' + menu['price']

    # get rid of 'item_id' and 'price' columns:

menu = menu[['item_price', 'item_category']]

menu.head()
```

```python
# Replace 'Smoothie' and 'Fresh Juice' columns with 'smoothie_juice' for the sake of random item selection:

menu.item_category.replace(['Smoothie', 'Fresh Juice'],
                           'smoothie_juice', 
                           inplace = True)
```

```python
# Create empty DataFrame for future storage: 

trans_wide = pd.DataFrame(columns = ['trans_id', 'item_price1', 'item_price2', 'item_price3', 'item_price4',
                      'item_price5', 'item_price6', 'item_price7', 'item_price8'])

long_list = [(str(x)) for x in list(range(7500))]

for i in long_list:

    # Create new empty list for each unique transaction:
    transaction = []

    # Generate unique transaction_id and append to list 'transaction': 
    id = 'trans_' + i
    transaction.append(id)

    # Iterate over item category and use probability to determine skip or select 1 or 2 items:
    for category in menu.item_category.unique():

        # Create category based subset of menu items
        to_choose = list(menu.loc[menu['item_category']== category, 'item_price'])

        # Select random number for probability sake:
        a = random.choice(range(1,100))

        if a >= 40: # customer orders 1 item
            item1 = random.choice(to_choose) 
            transaction.append(item1)
        elif a > 15:    # customer orders 2 items from the category
            item1 = random.choice(to_choose)
            item2 = random.choice(to_choose)
            transaction.append(item1)
            transaction.append(item2)
        else: # customer doesn't order from this category
            continue
    
    # Make all lists of length 9 by padding with np.nan
    transaction += [None] * (9 - int(len(transaction)))
    
    # Add transaction list to 'transaction_wide' DataFrame:
    trans_wide.loc[i] = transaction
```

```python
# Melt data to 2 columns: trans_id and item_price
df = trans_wide.set_index(keys = 'trans_id')
trans_wide.head()
```
